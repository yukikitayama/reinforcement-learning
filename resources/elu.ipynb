# Exponential linear unit

```
## setup

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
```

```
## exponential linear unit

def elu(x, alpha):
    if x >= 0:
        ans = x
    elif x < 0:
        ans = alpha * (np.exp(x) - 1)
    return ans
```

```
## test elu

alpha = 1
x = np.linspace(-5, 5, 100)
y = [elu(e, alpha) for e in x]

plt.plot(x, y, color = '#1f77b4')
plt.title(f'ELU with alpha {alpha}')
plt.xlabel('Activation input')
plt.ylabel('Activation output')
plt.axhline(y = 0, linestyle = '--', color = '#ff7f0e')
plt.axvline(x = 0, linestyle = '--', color = '#ff7f0e')
plt.axhline(y = -alpha, linestyle = '--', color = '#2ca02c')
plt.show()
```
